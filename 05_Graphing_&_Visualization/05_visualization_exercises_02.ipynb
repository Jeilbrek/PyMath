```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [-5, 5]
m = [0.7, -1.25]
b = [-2, .75]

for i in range(0, len(m)):
    y = m[i]*np.array(x)+b[i]
    plt.plot(x, y, label='y=%sx+%s' %(m[i], b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()
plt.title('The plot')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = range(-1, 4)
x = np.linspace(-4, 4, 100)

for i in e:
    y = x**i
    
    # Using the $ sign LaTeX code can be used
    plt.plot(x, y, label='$y=x^{%s}$' %i, linewidth=4)

    
plt.xlim(x[0], x[-1])
plt.ylim(-20,20)
plt.legend()    
plt.grid()
plt.show()
```
