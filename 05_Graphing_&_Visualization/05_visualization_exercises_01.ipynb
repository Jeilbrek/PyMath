```python
import matplotlib.pyplot as plt
import sympy as sym
```

```python
x = sym.symbols('x')
y = x**2-3*x

for xi in range(-11, 12):
    plt.plot(xi, y.subs(x, xi), 'o')

plt.xlabel('x')
plt.ylabel('f(x)=%s' %y)
plt.show()
```

```python
from math import sqrt

x = range(-21, 21)

for xi in x:
    plt.plot([0, xi], [0, sqrt(abs(xi))])

    
plt.show()
```
