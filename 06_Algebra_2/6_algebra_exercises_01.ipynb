```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.arange(1,6)

f1 = sum(a)/np.sum(a**2)
f2 = 1/sum(a)

f3 = np.prod(a)/np.prod(a**2)
f4 = 1/np.prod(a)

print(f1==f2)
print(f3==f4)
```

# Lecture 65

```python
x = np.linspace(-2,2,101)
f = x**2
df = np.diff(f)
dx = np.diff(x)


plt.plot(x,f)
plt.plot(x[:-1],df/dx)
plt.legend(['f(x)=$x^2$', '$\\frac{df}{dx}$'])
plt.grid()
plt.xlim(-1.8,1.8)
plt.ylim(-1,2)
plt.show()
```
